**lets make our first ANN using tensorflow**

```python
# Import necessary libraries
import tensorflow as tf
from keras import layers
from keras import Sequential
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, LabelEncoder

# Load the dataset
df = sns.load_dataset("tips")
print(df.head())  # Display the first few rows of the dataset

# Initialize the LabelEncoder
le = LabelEncoder()

# Encode categorical variables into numerical values
df["sex"] = le.fit_transform(df["sex"])
df["smoker"] = le.fit_transform(df["smoker"])
df["day"] = le.fit_transform(df["day"])
df["time"] = le.fit_transform(df["time"])

# Define features and target variable
X = df.drop("tip", axis=1)  # Features (input variables)
y = df["tip"]                # Target variable (output)

# Scale the features to the range [0, 1]
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)  # Fit and transform the features

# Split the dataset into training and testing sets
X_scaled_train, X_scaled_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.20, random_state=42
)

# Create the neural network model
model = Sequential([
    layers.Dense(32, activation='relu', input_shape=[X.shape[1]]),  # Input layer
    layers.Dense(16, activation='relu'),                             # Hidden layer
    layers.Dense(1)                                                  # Output layer
])

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
model.fit(X_scaled_train, y_train, epochs=50, batch_size=35)

# Evaluate the model on the test set
loss = model.evaluate(X_scaled_test, y_test)

# Print the loss
print('Loss:', loss)
```

```python
import tensorflow as tf
from keras import layers
from keras import Sequential
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
df=sns.load_dataset('tips')
df.head()
X=df.drop('tip',axis=1)
y=df['tip']
le=LabelEncoder()
X['sex']=le.fit_transform(X['sex'])
X["smoker"] = le.fit_transform(X["smoker"])
X["day"] = le.fit_transform(X["day"])
X["time"] = le.fit_transform(X["time"])
scaler=MinMaxScaler()
X_scaled=scaler.fit_transform(X)
X_scaled_train,X_scaled_test,y_train,y_test=train_test_split(X_scaled,y,test_size=0.20,random_state=42)
model=tf.keras.Sequential([
    layers.Dense(32,activation='relu',input_shape=(X_scaled.shape[1],)),
    layers.Dense(16,activation='relu'),
    layers.Dense(1)
])
model.compile(optimizer='adam',loss='mean_squared_error')
model.fit(X_scaled,y,epochs=50,batch_size=30)
loss=model.evaluate(X_scaled_test,y_test)
#model.predict(X_scaled_test,y_test)
print("loss",loss)
```
